```python
#human labels

import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa
import numpy as np
```

```python
df = pd.read_csv("google_form_responses.csv", index_col=False)

df = df.drop(columns=["What visual cues did you use to determine the person’s gender presentation?\n(e.g., clothing, hairstyle, body language)"])
df = df.drop(columns=["Email Address", "What gender do you identify yourself with?"])

df = df.T

df = df.reset_index(drop=True)
df = df.drop(index=0)
df = df.reset_index(drop=True)

df = df.replace({
    "Male-presenting": 0,
    "Female-presenting": 1,
    "Androgynous / Ambiguous": 2,
    "Cannot tell": 2
})

print(df.head())
```

```python
# Count how many raters voted 0 and 1 per image
category_counts = []

for _, row in df.iterrows():
    counts = [
        (row == 0).sum(),
        (row == 1).sum(), 
        (row == 2).sum()
    ]
    category_counts.append(counts)

category_counts = np.array(category_counts)

kappa = fleiss_kappa(category_counts)

print("Fleiss' Kappa (Human-Human):", kappa)
```
