```python
import pandas as pd
```

```python
df_dalle = pd.read_csv("gender_labels_ChatGPT_faces.csv", index_col=False)
df_dalle['Image'] = df_dalle['Image'].str.replace('.PNG', '_dalle', case=False)
df_midjourney = pd.read_csv("gender_labels_Midjourney_faces.csv", index_col=False)
df_midjourney['Image'] = df_midjourney['Image'].str.replace('.PNG', '_midjourney', case=False)
df_stable_diffusion = pd.read_csv("gender_labels_Stable Diffusion_faces.csv", index_col=False)
df_stable_diffusion['Image'] = df_stable_diffusion['Image'].str.replace('.PNG', '_stable_diffusion', case=False)

professions = pd.read_csv("professions_models.csv", index_col="id")
```

```python
combined_models = pd.concat([df_dalle, df_midjourney, df_stable_diffusion], axis=0)
combined_models = combined_models.reset_index(drop=True)

combined_professions = pd.concat([professions, combined_models], axis=1)

combined_professions.to_csv("final_ai_labels.csv", index=False)
```
